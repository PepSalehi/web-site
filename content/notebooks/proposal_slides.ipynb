```python
import matplotlib.pyplot as plt

def maybe_split_text(text):
    if len(text) <= 20:
        return text
    texts = text.split(" ")
    new_text = ""
    for t in texts:
        if len(new_text.split("\n")[-1] + t) <= 20:
            new_text += t + " "
        else:
            new_text += "\n" + t + " "
    return new_text[:-1]

def write_text_on_grid(ax, texts):
    if len(texts) > 11: # two columns
        texts = map(maybe_split_text, texts)
        ax.text(.25, .5, "\n".join(texts[:int(len(texts)/2)]), ha='center', va='center', fontsize=12)
        ax.text(.75, .5, "\n".join(texts[int(len(texts)/2):]), ha='center', va='center', fontsize=12)
    else:
        ax.text(.5, .5, "\n".join(texts), ha='center', va='center', fontsize=12)

def plot_innovation_grid(innovs, title):
    """
    """
    fig, axes = plt.subplots(2, 2, figsize=(9,12), sharex=True, sharey=True)
    font_props = dict(size=14)
    
    # I
    ax = axes[0,0]
    ax.grid(False)
    ax.set_xticks([])
    ax.set_yticks([])
    
    ax.text(.95, 0.01, "MARKET/INDIVIDUAL", fontdict=font_props, ha='right')
    write_text_on_grid(ax, innovs[0])
    
    # II
    ax = axes[0,1]
    ax.text(0.05, 0.01, "MARKET/NETWORK", fontdict=font_props, ha='left')
    write_text_on_grid(ax, innovs[1])
    
    # III
    ax = axes[1,0]
    ax.text(.95, .95, "NON-MARKET/INDIVIDUAL", fontdict=font_props, ha='right')
    write_text_on_grid(ax, innovs[2])
    
    # IV
    ax = axes[1,1]
    ax.text(0.05, .95, "NON-MARKET/NETWORK", fontdict=font_props, ha="left")
    write_text_on_grid(ax, innovs[3])
    
    fig.suptitle(title, y=0, fontsize=14)
    fig.tight_layout(pad=.5)
    fig.subplots_adjust(bottom=.02)
    return fig
```

## Tf-Idf

* Assign a weight to each term in a document that depends on **(1)** The term frequency $tf_{t,d}$: the number of occurrences of the term in a document

$$tf_{i,j}=\frac{n_{i,j}}{\sum_kn_{k,j}}$$

and **(2)** The document frequency $df_{t,d}$: the number of documents in which the term occurs

$$idf_i = \log\frac{|D|}{|\{d : t_i \in d \}|}$$

* The tf-idf is the term frequency scaled by the document frequency, often defined as

$tfidf(t,d) = log(tf_{t,d} +1)\times log\left( \frac{\|corpus\|}{df_{t}}\right)$

TF are corpus-wide normalized per-term observations counts
IDF are the negative logs of the corpus-wide normalized per-term document occurrence counts

Like the tf-idf but the IF are taken across the whole corpus

* Variants - tf - natural, logarithm, augmented, boolean, log ave
             df - none, idf, prob idf
             normalization - none, cosine, pivoted unique, byte size
cf. IR book, Chapter 6, Figure 6.15, p 128

## Feature Selection

* Need to identify predictive signals in the input features
* Two inter-related steps
1) How to choose your vocabularly
2) How to choose the words that help identify the decision boundaries between classes
* Helps avoid over-fitting - rare terms in the training data that only occur in one class might lead to new test data always being assigned to that class. It's a *noise feature*
* Should be important in difficult, high-skew situations
* Use some measure $A(t,c)$ for each term of the vocabulary and select the $k$ terms that have the highest values
* One-sided metrics (Odds ratio, Correlation coefficient, etc.)
  * related the features of a class with a sign, positive or negative
* Two-sided metrics (Chi-squared, Information Gain, etc.)
  * relate features without a sign, all features are considered equal

## Multi-Class Feature Selection

* Most of these are for binary class feature selection
* 1-of-N problems, where each document belongs to 1 and only 1 class out of N, can be solved by applying each
* Each task is solved independently - $class_i$ vs $~class_i$
* Since the estimators / induction algorithms often decompose the problem into n binary tasks and then make a final decision by some form of voting, it can be natural to select the features independently like this
* However, some 1-of-n induction algorithms do not perform binary decompositions and need *multi-class feature selection*
* Cf. Forman, 2004, A Pitfall and Solution in Multi-Class Feature Selection for Text Classification

## Information Gain Feature Selection

* Mldenic (1999), Yang and Pederson (1997)
* number of bits of information obtained for class prediction by knowing the presence or absence of a term in a document
* The information gain for term $t$, $G(t)$ over $i=[1,\dots,m]$ classes is defined to be

$$G(t) = -\sum_i^m P_r(c_i)\log P_r(c_i) + P_r(t)\sum_{i=1}^m P_r(c_i|t)\log P_r(c_i|t) + P_r(\bar{t})\sum_{i=1}^mP_r(c_i|\bar{t})\log P_r(c_i|\bar{t})$$

## Mutual Information Feature Selection

* Measure of how much information the presence/absence of a term contributes to making the classification decision on $c$.

$$I(U;C)=\sum_{e_t\in\{1,0\}}\sum_{e_c\in \{1,0\}}P(U=e_t, C=e_c)\log\frac{P(U=e_t,C=e_c}{P(U=e_t)P(U=e_c)}$$

* Where $U$ is a random variable $e_t = 1,0$ depending on whether the document contains term $t$ or not
* And $C$ is a random variable $e_c=1,0$ dependong on whether the document is in class $c$ or not
* We can use MLE for $P(\cdot)$
* MI is between 0 and 1. 0 indicates the the terms distribution within a class is the same as within the collection as a whole. 1 indicates the term is present in a document iff the document is in the class
* Yang and Pederson (1997) reported that MI performed terribly in multi-class problems

## $\chi^2$ Feature Selection

* In statistics the $chi^2$ measure is used to test for independence of two events
* For Feature selection, we would like to know if terms and document class occurrence are independent


$$^2(\mathbb{D},t,c)=\sum_{e_t\in \{0,1\}}\sum_{e_c\in \{0,1\}}\frac{(N_{e_te_c}-E_{e_te_c})^2}{E_{e_te_c}}$$

* Where $e_t$ and $e_c$ are the same as with MI and $N$ is the observed frequency in $\mathbb{D}$ while $E$ is the expected frequency
* Potential problems
1. one degree of freedom so we apply the Yates correction
2. multiple comparisons problem

## Multinomial Naive Bayes


* A probabilistic learning method
* The probability of a document $d$ being in class $c$ is computed as

$P(c|d) \propto P(c) \prod_{1 \leq k \leq n_d} P(t_k|c)$

where $P(t|k)$ is the conditional probablity of term $t_k$ occurring in a document of class $c$.

* $P(t|k)$ can be interpreted as a measure of how much evidence $t_k$ contributes that $c$ is the correct class
* P(c) is the prior probability of a document belonging to class $c$
* If there is no evidence in the data, then the document gets assigned according to the prior
* We are seeking *maximum a posteriori* (MAP) estimates for class $c$

$$c_{map}=\underset{c\in \mathcal{C}}{\arg \max}\hat{P}(c)\prod_{1 \leq k \leq n_d}\hat{P}(t_k|c)$$

## Estimation of MNB

* We work with the log of the MAP estimator for numerical reasons
* We start with the MLE estimates for $\hat{P}(c)$ and $\hat{P}(t|c)$
* The MLE for the prior is the frequency $\hat{P}(c) = \frac{N_c}{N}$
* Likewise the MLE for the conditional probability is $\hat{P}(t|c)=\frac{T_{ct}}{\sum_{t^\prime \inV}T_{ct^\prime}}$
* The problem with the MLE estimates is that it is zero for a term-class co-occurrence that's not in the training data set
* This is an example of sparseness, the training data is never big enough to represent the frequency of rare events
* To eliminate the zeros, we use *add-one* or *Laplace smoothing*, which adds one to each count

$$\hat{P}(t|c)=\frac{T_{ct}+1}{\sum_{t^\prime \in V}(T_{ct^\prime}+1)}=\frac{T_{ct}+1}{\sum_{t^\prime \in V}T_{ct^\prime}+B}$$

where $B=|V|$
* This can be interpreted as a uniform prior for each term

* *positional independence* - bag-of-words

## Support vector machines

Boser, B.E., Guyon, I. and Vapnik, V. 1992 "A training algorithm for optimal margin classifiers," in *COLT*
Joachims, T. "Text categoirzation with support vector machines: learning with many relevant features," in *ECML*, 1998. pp. 137-142.

* Given a training features vector $x_i \ in \mathbb{R}^n$ and labels $y_i \in {-1, 1}, i=1,\dots,l$ SVM solves the following problem

$$\min_w\frac{1}{2}\parallel \boldsymbol{w} \parallel + C\sum_{i=1}^l\max(0, 1-y_i\boldsymbol{w}^T\boldsymbol{x}_i)$$

* C > 0 is the regularization parameter.
* The second term is the sum of training losses
* Given a new test instance $\boldsymbol{x}$, the predicted label is sgn(\boldsymbol{w}^Tx)
* Obviously, this is only appropriate for binary classification
* To solve multi-class problems, it is common to use the one-versus-rest method
* Given $k$ classes, train $k$ models solving for $\boldsymbol{w}_1,\cdots,\boldsymbol{w}_k$. 
* The solutions to these models, as the name suggests, are for each class $k$ being positive and not in $k$ being negative
* For an instance $\boldsymbol{x}$ with label $y$, we have

$$(w^y)^Tx \ge 1 \text{and} (w^m)^Tx \le -1 \forall m \neq y$$

The predicted label is

$$\underset{m=1,\dots,k}{argmax}(w^m)^Tx$$

Other approaches
1. one-vs-one multi-class method
2. multi-class SVM - Crammer and Singer

## Measure of Performance

* F1 as a function of vocabulary size for the different estimators
$F_B=\frac{(B^2+1)PR}{B^2P+R}$ <br />
B=1 indicates that precision and recall are of equal importance. B=2 recall is  twice as important as precision. And B=.5, recall is half as important as precision.

**recall**
$$R=\frac{a}{a+b}$$
**precision**
$$P=\frac{a}{a+c}$$
where $a$ is the number of correctly classified positive documents, b is the number of incorrectly classified positive documents, and $c$ is the number of incorrectly classified negative documents.<br />
**accuracy** <br />
**single measure that combines individual classifiers** <br />
    microaveraging - pool per-document decisions across classes then computes effectiveness measure on the pooled contingency table
    macroaveraging - single average over classes

* in classification, precision is the number of true positives (number of documents correctly labelled as belonging to the positive class) divided by the total number of elemens labeled as belonging to the positive class (true and false positives)
* in statistical terms, maximum precision is equivalent to the absence of type I errors
* in classification, recall is defined as the number of true positives divided by the total number of elements that actually belong to that class
* in statistical terms, maximum recall is equivalent to the absence of type II errors

* $\rho$ is the spatial autocorrelation parameter
* $\theta$ marginal impact on dependent variable from neighbors covariates
